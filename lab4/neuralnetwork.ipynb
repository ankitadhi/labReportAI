```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = 1.5
y= 0.5
```

```python
W =np.random.randn(1)
print(W, W.shape)
```

```python
#training Loop
epochs = 20
lr=0.1
eList =[]
for epoch in range(epochs):
    #forward pass
    y_out = X*W

    #compute the error
    error = np.square(y-y_out)
    eList.append(error)

    #compute the gradient
    gradient = -2*(y-y_out)*X

    #back propagate the error
    W= W-gradient*lr



```

```python
plt.plot(range(epochs), eList)
plt.xlabel('Epoch')
plt.ylabel('Error')
plt.title('Epoch vs Error')
```

```python
def output(X, W):
    return X*W
print(output(X, W))
```
