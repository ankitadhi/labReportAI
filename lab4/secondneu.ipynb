```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([[1.5, 0.33]])
y =0.65
print(X.shape)
```

```python
W = np.random.randn(1,2)
print(W, W.shape)
```

```python
y_out = np.dot(X, np.transpose(W)) #taking the dot product
print(y_out, y_out.shape)
```

```python
#training loop
epochs = 20
lr = 0.1
eList = []
for epoch in range(epochs):
    #forward pass
    y_out = np.dot(X, np.transpose(W)) #taking the dot product

    #error compute
    error = np.square(y-y_out)
    eList.append(error[0][0])

    gradient = -2*(y-y_out)*X

    W = W-gradient*lr




```

```python
plt.plot(range(epochs), eList)
plt.xlabel('Epoch')
plt.ylabel('Error')
plt.title('Epoch vs Error')
```

```python
def output(X, W):
    return np.dot(X, np.transpose(W))


print(output(X, W))
```
