```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = np.random.randn(100)

# Compute basic statistics
mean = np.mean(data)
std_dev = np.std(data)

print(f"Mean: {mean:.2f}, Standard Deviation: {std_dev:.2f}")
```

```python
# Import the required library
import matplotlib.pyplot as plt

# Data for plotting
x = [0, 1, 2, 3, 4, 5]  # X-axis values
y = [0, 1, 4, 9, 16, 25]  # Y-axis values (squared numbers)

# Create a plot
plt.plot(x, y, label='y = x^2')  # Plot x vs y

# Add labels and title
plt.xlabel('X-axis')  # X-axis label
plt.ylabel('Y-axis')  # Y-axis label
plt.title('Simple Plot of y = x^2')  # Title of the plot

# Show grid for better readability
plt.grid(True)

# Show the legend
plt.legend()

# Display the plot
plt.show()
```
