```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

```

```python
m = 0.7
c = 0.3
X= np.arange(0,1,0.02)

y =m*X +c

print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=10)
```

```python
def plot_data(X_train = X_train, X_test=X_test, y_train=y_train, y_test=y_test, predictions=None):
    plt.figure(figsize=(8,6))
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('x vs Y')
    plt.scatter(X_train, y_train, c='b', s=8, label = 'Training Data')
    plt.scatter(X_test, y_test, c='r', s=8, label='Testing Data')
    if predictions is not None:
        plt.scatter(X_test, predictions, c='g',  s=8, label= 'testing data')
plot_data()
```
